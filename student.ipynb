```python
# Strip whitespace from column names
df.columns = df.columns.str.strip()
df
```

```python
# Remove leading/trailing spaces in string columns
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
df
```

```python
# Drop rows where key identifying info is missing
df = df.dropna(subset=['Event.Id', 'Make', 'Model'])
df
```

```python
# Fill remaining missing values in numeric columns with 0
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].fillna(0)
df
```

```python
# Fill missing values in object (string) columns with 'Unknown'
object_cols = df.select_dtypes(include=['object']).columns
df[object_cols] = df[object_cols].fillna('Unknown')
df
```

```python
# remove duplicate rows if any ---
df = df.drop_duplicates()
df
```

```python
# --- Convert Event.Date to datetime ---

df['Event.Date'] = pd.to_datetime(df['Event.Date'], errors='coerce')
df
```
